```python
import json
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def graf_Anex_G (d, col, tit, ylab):
    dicc_data= d[col]
    final_df = pd.DataFrame([[key, dicc_data[key]] for key in dicc_data.keys()], columns=['date', col])
    final_df["date"]= pd.to_datetime(final_df["date"], format = "%Y-%m-%d")
    final_df.set_index("date", inplace=True)
    plt.figure()
    plt.plot()
    final_df[col].plot(kind = "line", title = tit , grid = False, figsize=(13,9)) 
    plt.xlabel ("Date") 
    plt.ylabel (ylab)
    plt.show()
```
